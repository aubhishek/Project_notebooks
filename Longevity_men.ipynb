#### This is finding median values on 50

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the percentiles and their corresponding values
percentiles = [5, 10, 25, 50, 75, 90, 95]
percentile_values = [29, 32.1, 40.1, 48, 55.2, 61.8, 66.3]

# Create a normal distribution based on the percentiles
mean = np.percentile(percentile_values, 50)  # Median value
std_dev = (np.percentile(percentile_values, 75) - np.percentile(percentile_values, 25)) / 1.349  # IQR/1.349
data = np.random.normal(mean, std_dev, 10000)  # Generating random data with the same distribution

# Create a histogram plot
plt.hist(data, bins=50, density=True, alpha=0.6, color='g', label='Generated Data')
plt.axvline(mean, color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(np.percentile(data, 25), color='b', linestyle='dashed', linewidth=2, label='25th Percentile')
plt.axvline(np.percentile(data, 75), color='y', linestyle='dashed', linewidth=2, label='75th Percentile')

# Plot the provided percentiles as vertical lines
for percentile, value in zip(percentiles, percentile_values):
    plt.axvline(np.percentile(data, percentile), color='k', linestyle='dotted', linewidth=2, label=f'{percentile}th Percentile: {value}')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Distribution with Percentiles')
plt.legend()

# Show the plot
plt.show()
```

#### Using ipywidget to show the shift in medians

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Define a function to plot the distribution for a given age group
def plot_distribution(i):
    percentiles = [5, 10, 25, 50, 75, 90, 95]
    percentile_values = data_as_np_array[i, :]

    mean = np.percentile(percentile_values, 50)
    std_dev = (np.percentile(percentile_values, 75) - np.percentile(percentile_values, 25)) / 1.349
    data = np.random.normal(mean, std_dev, 10000)

    plt.figure(figsize=(8, 6))
    plt.hist(data, bins=50, density=True, alpha=0.6, label=f'Age group {rows[i]}')
    
    for percentile, value in zip(percentiles, percentile_values):
        plt.axvline(np.percentile(data, 50), color='k', linestyle='dotted', linewidth=2)
    
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.title(f'Distribution for Age Group {rows}')
    plt.legend()
    plt.show()

# Create a slider for age groups
age_group_slider = widgets.IntSlider(value=0, min=0, max=data_as_np_array.shape[0] - 1, description='Age Group:')
widgets.interactive(plot_distribution, i=age_group_slider)
```
